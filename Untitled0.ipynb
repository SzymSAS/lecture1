```python
import pandas as pd
import matplotlib.pyplot as plt

# Poprawny link do surowych danych CSV
url = "https://raw.githubusercontent.com/SzymSAS/lecture1/main/Salary%20Data%20(2).csv"  # Podmień na prawdziwy RAW link

# Wczytanie danych
df = pd.read_csv(url)

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
plt.scatter(df["Years of Experience"], df["Salary"], color='blue', alpha=0.6)
plt.xlabel("Lata doświadczenia")
plt.ylabel("Wynagrodzenie")
plt.title("Zależność wynagrodzenia od lat doświadczenia")
plt.grid(True)

# Wyświetlenie wykresu
plt.show()
```
